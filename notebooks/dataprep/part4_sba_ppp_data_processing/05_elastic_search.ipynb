```python
import elasticsearch
import elasticsearch_dsl
import pandas as pd
```

```python
es = elasticsearch.Elasticsearch(['http://34.217.44.96:8082'])
```

```python
columns_suppliers = [
           'BorrowerName', 
           'NAICSCode', 
#            'RaceEthnicity', 
           'BusinessType', 
           'full_address', 
            'latitude',
            'longitude',
#            'postal_code', 
#            'place_name', 
#            'state_name', 
#            'urban_rural-population-total_population',
#             'urban_rural-percent-rural_population-of-total_population', 
           'location', 
           'RuralUrbanIndicator', 
           'BusinessAgeDescription', 
#            'NonProfit', 
#            'Veteran', 
#            'Gender', 
            'OfficialNAICSDescription',
            'urban_rural-percent-rural_population-of-total_population'
]
```

```python
request1 = (elasticsearch_dsl.Search(using=es, index='suppliers')
            .filter('term', **{'urban_rural-percent-rural_population-of-total_population': 1.0})
            .query("match", NAICSCode='326199'))

sample = request1[0:25]
df = pd.DataFrame([dict((c, r[c]) for c in columns_suppliers) for r in sample])
df
```

```python
request1 = (elasticsearch_dsl.Search(using=es, index='suppliers')
            .query("match", NAICSCode='326199')).query("range", **{'urban_rural-percent-rural_population-of-total_population': {"gte" : 0.0, "lt" : 1.0}} )
            

sample = request1[0:10000]
df = pd.DataFrame([dict((c, r[c]) for c in columns_suppliers) for r in sample]).sort_values(by='urban_rural-percent-rural_population-of-total_population', ascending=False)
df.head()
```

```python
df_all = pd.DataFrame(columns = columns_suppliers)
for n in list(['326199', '236220']):
    request1 = (elasticsearch_dsl.Search(using=es, index='suppliers')
                                .filter('terms', RuralUrbanIndicator=['R'])
                                .query("match", NAICSCode=n) # Will also work                                
              )
    request2 = (elasticsearch_dsl.Search(using=es, index='suppliers')
                                .filter('terms', RuralUrbanIndicator=['U'])
                                .query("match", NAICSCode=n) # Will also work                                
              )
    sample = request1[0:25]
    df = pd.DataFrame([dict((c, r[c]) for c in columns_suppliers) for r in sample])
    df_all = df_all.append(df)
    
    sample = request2[0:25]
    df = pd.DataFrame([dict((c, r[c]) for c in columns_suppliers) for r in sample])
    df_all = df_all.append(df)
    print(len(df))
```

```python
df_all.reset_index(drop=True)
```
